```python
import numpy as np 
import pandas as pd 
```

```python
GROUP BY
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
#give or groupby sum of sales of A,B
dat = df.groupby('Category')
dat
for key,values in dat:
    print(key)
    print(values)
    
```

```python
#give or groupby sum of sales of A,B gives total sum of sales by catagory
dat = df.groupby('Category')['Sales'].sum()
dat
```

```python
# Group by Store and calculate the sum of Sales
dfa= df.groupby('Store')['Sales'].sum()
dfa
```

```python
# Group by multiple columns
dff = df.groupby(['Category','Store'])['Sales'].sum()
dff
```
